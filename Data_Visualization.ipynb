```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("car_sales_data.csv")

# Quick view
print(df.head())
```

```python
#LINE CHART
plt.figure(figsize=(10,6))
plt.plot(df["Year of manufacture"], df["Price"], marker='o', linestyle='-', color='b')
plt.title("Car Price Trend Over Years")
plt.xlabel("Year of Manufacture")
plt.ylabel("Price")
plt.show()
```

```python
#BAR CHART
avg_price = df.groupby("Manufacturer")["Price"].mean()

plt.figure(figsize=(10,6))
avg_price.plot(kind='bar', color='skyblue')
plt.title("Average Car Price by Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
#HISTOGRAM
plt.figure(figsize=(10,6))
plt.hist(df["Mileage"], bins=20, color="orange", edgecolor="black")
plt.title("Distribution of Car Mileage")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()
```

```python
#PIE CHART
fuel_counts = df["Fuel type"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
#BOX PLOT
plt.figure(figsize=(10,6))
df.boxplot(column="Price", by="Fuel type")
plt.title("Car Price Distribution by Fuel Type")
plt.suptitle("")  # removes the default extra title
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```

```python
#HEATMAP
# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Car Features")
plt.show()
```

```python
#VIOLIN PLOT
plt.figure(figsize=(10,6))
sns.violinplot(x="Fuel type", y="Price", data=df, palette="Set2")
plt.title("Car Price Distribution by Fuel Type (Violin Plot)")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```

```python
#LINE PLOT
# Group by year and take average price
yearly_price = df.groupby("Year of manufacture")["Price"].mean()

plt.figure(figsize=(10,6))
plt.plot(yearly_price.index, yearly_price.values, marker="o", color="b", linestyle="-")
plt.title("Average Car Price Over Years")
plt.xlabel("Year of Manufacture")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

```python
#POINT PLOT
plt.figure(figsize=(8,6))
sns.pointplot(x="Fuel type", y="Price", data=df, estimator="mean", ci=None, color="blue", markers="o", linestyles="-")
plt.title("Average Car Price by Fuel Type (Point Plot)")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

```python
#KDE PLOT
plt.figure(figsize=(8,6))
sns.kdeplot(df["Price"], shade=True, color="blue")
plt.title("KDE Plot of Car Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```
